```python
import pandas as pd
df=pd.read_csv('Task (2) Dataset.csv')
df.info()
```

# Tasks

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
import matplotlib.pyplot as plt
df = pd.DataFrame(df)
df.plot.hexbin(x='JamsCount', y='JamsCount', title="hixbin Plot of Traf fic Index Week Ago vs Jams Count " , gridsize=15)
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.histplot(df['TravelTimeLive'], kde=True, bins=10 )
plt.title('Histogram of Travel Time Live')
plt.xlabel('Travel Time Live')
plt.ylabel('Fequency')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
import pandas as pd
import matplotlib.pyplot as plt
# Histogram for TrafficIndexLive
plt.figure(figsize=(10, 6))
plt.hist(df['TrafficIndexLive'], bins=20, edgecolor='black')
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
#Histogram for Datetime
plt.figure(figsize=(12, 6))
df['Datetime'] = pd.to_datetime(df['Datetime'])
plt.hist(df['Datetime'], bins=40, edgecolor='black')
plt.title('Distribution of Datetime')
plt.xlabel('Datetime')
plt.ylabel('Frequency')
plt.show()
#Histogram for TravelTimeLive
plt.figure(figsize=(12, 6))
df['TravelTimeLive'] = pd.to_datetime(df['TravelTimeLive'])
plt.hist(df['TravelTimeLive'], bins=30, edgecolor='black')
plt.title('Distribution of Travel Time Live')
plt.xlabel('Travel Time Live')
plt.ylabel('Frequency')
plt.show()

```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
#line chart for JamsLengthvs TrafficIndexLive
plt.figure(figsize=(12, 6))
sns.lineplot(x='JamsLength', y='TrafficIndexLive', data=df)
plt.title('Traffic Index vs. Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Traffic Index')
plt.show()
#line chart for TravelTimeLive vs TravelTimeHistoric
plt.figure(figsize=(12, 6))
sns.lineplot(x='TravelTimeLive', y='TravelTimeHistoric', data=df)
plt.title('Traffic Index vs. Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Traffic Index')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JamsDelay', y='TrafficIndexLive', data=df)
plt.title('Jams Delay vs Traffic Index Live')
plt.xlabel('JamsDelay ')
plt.ylabel('TrafficIndexLive')
plt.show()


```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
import matplotlib.pyplot as plt
# Bar chart for Traffic Index by City
plt.figure(figsize=(7, 6))
sns.barplot(x='City', y='TrafficIndexLive', data=df)
plt.title('Traffic Index by City')
plt.xlabel('City')
plt.ylabel('Traffic Index')
plt.xticks(rotation=45)
plt.show()

# Pie chart for Traffic Index distribution
traffic_index_counts = df['TrafficIndexLive'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(traffic_index_counts, labels=traffic_index_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Traffic Index Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['JamsCount'], df['TrafficIndexLive'])
plt.title('Jams Count vs Traffic Index Live')
plt.xlabel('Jams Count')
plt.ylabel('Traffic Index Live')
plt.show()


```

```python
# Heatmap
import numpy as np
df = df[['Datetime','TrafficIndexLive','JamsCount','JamsDelay','JamsLength','TrafficIndexWeekAgo','TravelTimeLive']].apply(pd.to_numeric,errors = 'coerce')
corr = df.corr()
sns.heatmap(corr, annot=True)
```
